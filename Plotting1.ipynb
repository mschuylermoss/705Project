# Q 1 b - Also DONE!

```python
Ts = np.arange(1.0,4.01,0.01)
chains = np.arange(1,11,1)
Ls = [8,12,16,20,24]
J = 1

energies = dict()
rawMs = dict()
Ms = dict()
CV = dict()
XT = dict()

for L in Ls:
    energies_list = []
    rawMs_list = []
    Ms_list = []
    CV_list = []
    XT_list = []
    print(f"L = {L}")
    for T in Ts:
        T_st = round(T,2)
        chains_e_list = []
        chains_rawM_list = []
        chains_M_list = []
        chains_CV_list = []
        chains_XT_list = []
        for chain in chains:
            df = pd.read_csv(f'./data/q1b_cluster/L_{L}/avgvals_T_{T_st}_chain{chain}.csv')
            energies_np = np.array(df["avenergy"])
            energies2_np = np.array(df["avenergy2"])
            CV_val = (energies2_np[-1] - (energies_np[-1])**2)/T**2
            rawMs_np = np.array(df["rawMs"])
            Ms_np = np.array(df["avM"])
            Ms2_np = np.array(df["avM2"])
            XT_val = (Ms2_np[-1] - Ms_np[-1]**2)/T

            chains_e_list.append(energies_np[-1])
            chains_rawM_list.append(rawMs_np[-1])
            chains_M_list.append(Ms_np[-1])
            chains_CV_list.append(CV_val)
            chains_XT_list.append(XT_val)
            
        energies_list.append(np.mean(chains_e_list))
        rawMs_list.append(np.array(chains_rawM_list))
        Ms_list.append(np.mean(chains_M_list))
        CV_list.append(np.mean(chains_CV_list))
        XT_list.append(np.mean(chains_XT_list))
        
    energies[f"{L}"] = energies_list
    rawMs[f"{L}"] = rawMs_list
    Ms[f"{L}"] = Ms_list
    CV[f"{L}"] = CV_list
    XT[f"{L}"] = XT_list
```

```python
fig = plt.figure(1,figsize=(7,3), dpi=120, facecolor='w', edgecolor='k')
criticaltemp = 2/np.log(1+np.sqrt(2))
colors = plt.cm.Greens(np.linspace(0, 1, 4*len(Ls)))[len(Ls):]
for i in range(len(Ls)):
    L = Ls[i]
    plt.scatter(Ts,np.array(energies[f"{L}"])/L**2,s=5,color=colors[3*i],label=f'L = {L}')
plt.ylabel('$\\langle E \\rangle /N$',fontsize=15)
plt.xlabel('T',fontsize=15)
plt.vlines(criticaltemp,-2,-0.4,linestyle='--',color='k',label="$T_{c}$")
plt.legend()
plt.show()
```
